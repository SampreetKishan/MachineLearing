### Objective: The company is trying to decide whether to focus their efforts on their mobile app experience or their website 

```python
#Let's import the necessary modueles
import pandas as pd
import os 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Let's read the dataset
raw_dataset = pd.read_csv("ECommerceCustomers.csv")
```

## Exploratory Data Analysis 

```python
print(raw_dataset.head(5)) 
```

```python
#Let's try to clean up the dataset
print("Number of rows in the raw dataset: ", raw_dataset.shape[0])

#let's drop rows with missing values
dataset_cleaned = raw_dataset.dropna(axis='index')

print("\nNumber of rows in the cleaned dataset: ", dataset_cleaned.shape[0])

print("\nHere's some information about the dataset: ") 
dataset_cleaned.info()
```

```python
dataset_cleaned.describe()
```

### Correlation between the numerical columns 

```python
#Let's look at the correlation between certain columns 
corr_columns = ['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership', 'Yearly Amount Spent']
dataset_cleaned[corr_columns].corr().style.background_gradient(cmap='YlOrRd')
```

### Scatterplots to visualise correlation between different variables 

```python
#Let's use seaborn to create a scatterplot to show the correlation between the different columns
sns.pairplot(dataset_cleaned[corr_columns],kind='reg', plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha': 0.1}})
```

_We can see a positive correlation between Yearly amount spent and Time on App. However, there is a negative correlation between Yearly amount spent and Time on Website. However, it seems like Yearly amount spent is more closely correlated to the length of membership than it is to Time on App_

## Splitting the dataset into train and test datasets 

```python
#Let's build our linear model.
#But before we do that, let's split the data into training and testing data 
from sklearn.model_selection import train_test_split

print("Let's split the dataset into a dataset that contains features and another dataset for the y/target values..\n") 
x_columns = ['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']
y_column = ['Yearly Amount Spent']
X_data = dataset_cleaned[x_columns]
y_data = dataset_cleaned[y_column]

#print(X_data.head(5))

#print(y_data.head(5))

print("Let's split the datasets into training and testing datasets..")
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.33, random_state=42)

```

## Creating the model 

```python
from sklearn.linear_model import LinearRegression
print("Let's fit our X_train and y_train datasets to a Linear Regression model") 
print("\nCreating the model")
model = LinearRegression().fit(X_train, y_train)
print("\nR^2 value:",model.score(X_train, y_train))
print("The coefficients are: ", model.coef_) 
print("The intercept is: ", model.intercept_) 

```

## Testing the model 

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

#Let's run the model to get predicted values for y (yearly amount spent) for the testing data 
y_predicted = model.predict(X_test)
#y_predicted 

mse = mean_squared_error(y_test, y_predicted)
mae = mean_absolute_error(y_test, y_predicted)

print("The mean squared error: ", mse)
print("The mean absolute error: ", mae) 

y_test_array = y_test.values 

#Looks like both the y_predicted and y_test_array are arrays each and value is stored as an array 
# array([[404.06761931],
#       [542.58543915],
#       [427.34884845],
#        [502.01838314],.....

#We need to convert them into flat one-dimensional arrays 

y_predicted_flat = [item[0] for item in y_predicted]
y_test_flat = [item[0] for item in y_test_array ]

#Let's draw a scatterplot between y_test and y_predicted values 
sns.scatterplot(x=y_test_flat, y=y_predicted_flat)
plt.xlabel('Yearly spent amount values from TEST dataset')
plt.ylabel('Yearly spent amount PREDICTED values')
plt.title('Predicted vs Test values') 
plt.grid(True)
plt.show() 
```

### Analysis of the residuals 

_The objective is to make sure that the residuals form a normal(or close to normal) distribution to ensure there isn't any bias in your model_ 

```python
#Let's reconvert the lists to dataframes
y_test_flat = pd.DataFrame(y_test_flat)
y_predicted_flat = pd.DataFrame(y_predicted_flat)

residuals = y_test_flat - y_predicted_flat 
#residuals 

sns.displot(residuals, kde=True, bins=30)
```

### Looks like the residuals are pretty much in a normal distribution.
