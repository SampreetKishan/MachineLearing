```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

### _Ingesting the data and performing some exploratory analysis_

```python
raw_data = pd.read_csv("data.csv") 
print(raw_data.head(5))
```

```python
print("\n Some more information about the dataset: ", raw_data.info())
```

```python
#let's describe the data
raw_data.describe()
```

### _Cleaning the dataset_ 

```python
#You see that column "Unnamed: 32" is null so we are going to remove it
raw_data = raw_data.drop(['Unnamed: 32'], axis=1)

#We are also going to remove columns that don't add matter to analysis like "id"
raw_data = raw_data.drop(['id'], axis=1)

raw_data.info()
```

```python

# We know that the column diagnosis is our target column which should be a cateogorical variable so let's do that.
raw_data["diagnosis"] = raw_data["diagnosis"].astype('category')

#Let's convert Ms to 1 and Bs to 0 
mapping = {'M': 1, 'B': 0}
raw_data['diagnosis'] = raw_data['diagnosis'].replace(mapping)

raw_data.head(5)
```

### _Count of malignant and benign entries_

```python
#Let's plot the number of malignant and benign entries
value_counts = raw_data['diagnosis'].value_counts()

value_counts.plot(kind='bar', color='skyblue')  # Adjust color as desired
plt.title('Count of malignant and  benign entries')
plt.xlabel('1: Malignant & 0:benign')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Optional: Rotate x-axis labels for better readability
plt.show()
```

### _Some more clean up before building the model_ 

```python
#let's build our logistic regression model

#Let's first split the data into X_data(features) and y_data(target) 
X_data = raw_data.drop(['diagnosis'], axis=1)
y_data = raw_data['diagnosis']

#print(X_data.head(5))
#print(y_data.head(5)) 
```

### _Spliting the dataset into training and testing data_

```python
#Now let's split the data into X_train, X_test, y_train, y_test 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_data, y_data, test_size=0.33, random_state=42)
```

### _Now building the model using the traing data_

```python
#Let's build the logisitic regression model
from sklearn.linear_model import LogisticRegression

#Intantiate an object of class LogisticRegression
logR_model = LogisticRegression()

#Fit the X_train and y_train data to the logR_model
logR_model.fit(X_train, y_train)
```

```python
y_pred = logR_model.predict(X_test)

#print(y_pred)
#print(y_test)
```

### _Evaluating the model_

```python
# Evaluating the model 

from sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(y_test, y_pred, average='macro')
```

### _Precision, Recall F1_Score, and Support of the model against test data_ 

```python
precision, recall, f1_score, support = precision_recall_fscore_support(y_test, y_pred, average=None)

class_of_interest = 1


if(class_of_interest==1):
    malign_or_benign = 'MALIGN'
else:
    malign_or_benign = 'BENIGN'
    


class_precision = precision[class_of_interest]
class_recall = recall[class_of_interest]
class_f1 = f1_score[class_of_interest]
class_support = support[class_of_interest]

print(f"Precision for class {malign_or_benign}: {class_precision:.4f}")
print(f"Recall for class {malign_or_benign}: {class_recall:.4f}")
print(f"F1-score for class {malign_or_benign}: {class_f1:.4f}")
print(f"Support (number of instances) for class {malign_or_benign}: {class_support}")
```
